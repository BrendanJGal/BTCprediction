```python
from pandas_datareader import data
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.pylab import rcParams

from keras.models import Sequential
from keras.layers import LSTM,Dropout,Dense

from sklearn.preprocessing import MinMaxScaler
```

```python
btc_df = data.DataReader('BTC-USD', 'yahoo', '2020-08-01', '2021-12-01')
btc_df.reset_index(level=0, inplace=True)
btc_df.head()
```

```python
btc_df["Date"] = pd.to_datetime(btc_df.Date, format="%Y-%m-%d")
btc_df.index=btc_df['Date']

plt.figure(figsize=(16,8))
plt.plot(btc_df["Close"], label='Close Price')
```

```python
btc_close_data = btc_df.filter(['Close'])
btc_close_ds = btc_close_data.values
btc_training_data_leng = int(np.ceil(len(btc_close_ds)) * .95)

btc_training_data_leng
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_btc = scaler.fit_transform(btc_close_ds)

scaled_btc
```

```python
btc_train_data = scaled_btc[0:int(btc_training_data_leng), :]

btc_x_train = []
btc_y_train = []

for i in range(60, len(btc_train_data)) :
    btc_x_train.append(btc_train_data[i-60:i, 0])
    btc_y_train.append(btc_train_data[i, 0])
    if i <= 61:
        print(btc_x_train)
        print(btc_y_train)
        print()

btc_x_train, btc_y_train = np.array(btc_x_train), np.array(btc_y_train)

btc_x_train = np.reshape(btc_x_train, (btc_x_train.shape[0], btc_x_train.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (btc_x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(btc_x_train, btc_y_train, batch_size=1, epochs=1)
```

```python
btc_test_data = scaled_btc[btc_training_data_leng - 60:, :]

# Create the data sets x_test and y_test
btc_x_test = []
btc_y_test = btc_close_ds[btc_training_data_leng:, :]
for i in range(60, len(btc_test_data)): 
    btc_x_test.append(btc_test_data[i-60:i, 0])

# Convert the data to a numpy array
btc_x_test = np.array(btc_x_test)

# Reshape the data

btc_x_test = np.reshape(btc_x_test, (btc_x_test.shape[0], btc_x_test.shape[1], 1))

# Get the models predicted price values 
btc_predictions = model.predict(btc_x_test)
btc_predictions = scaler.inverse_transform(btc_predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((btc_predictions - btc_y_test) ** 2)))
rmse
```

```python
btc_train = btc_close_data[:btc_training_data_leng]
btc_valid = btc_close_data[btc_training_data_leng:]
btc_valid['Predicitons'] = btc_predictions
#btc_df["Date"] = pd.to_datetime(btc_df.Date, format="%Y-%m-%d")
#btc_df.index=btc_df['Date']

#plt.figure(figsize=(16,8))
#plt.plot(btc_df["Close"], label='Close Price')

plt.figure(figsize=(16,8))
#plt.xlabel('Date', fontsize=18)
#plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(btc_train["Close"])
plt.plot(btc_valid["Close"])
plt.plot(btc_valid["Predicitons"])
plt.legend(["Train", "Val", "Predictions"], loc='lower right')
plt.show()
```
